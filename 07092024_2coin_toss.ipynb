```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
coin = ["Heads", "Tails"]
```

```python
number_of_tosses = 100
```

```python

tosses1 = np.random.choice(coin, number_of_tosses)
tosses2 = np.random.choice(coin, number_of_tosses)

  #number of combinations (HH, HT, TH, TT)
nr_hh = np.sum((tosses1=='Heads')&(tosses2=='Heads'))
nr_ht = np.sum((tosses1=='Heads')&(tosses2=='Tails'))
nr_th = np.sum((tosses1=='Tails')&(tosses2=='Heads'))
nr_tt = np.sum((tosses1=='Tails')&(tosses2=='Tails'))
```

```python
#Probabilities
prob_head_head = nr_hh / number_of_tosses
prob_head_tails = nr_ht / number_of_tosses
prob_tails_head = nr_th / number_of_tosses
prob_tails_tails = nr_tt / number_of_tosses
```

```python
print(f'Probability of heads and heads: {prob_head_head}')
print(f'Probability of heads and tails: {prob_head_tails}')
print(f'Probability of tails and heads: {prob_tails_head}')
print(f'Probability of tails and tails: {prob_tails_tails}')
```

```python
import matplotlib.pyplot as plt
```

```python
x_values = ['head-head', 'head-tails', 'tails-head', 'tails-tails']
frequencies = [nr_hh, nr_ht, nr_th, nr_tt]

plt.bar(x_values, frequencies, color=['lightpink', 'skyblue', 'gold', 'palegreen'])
plt.title(f'Frequencies of each coin toss outcome, {number_of_tosses} throws')
plt.xlabel('Coin sides')
plt.ylabel('Number of throws')
```
